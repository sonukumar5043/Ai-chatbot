```python
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
print(gr.__version__)
```

```python
client = OpenAI(
    api_key="your api key",
    base_url="your api endpoint"
)
```

```python
system_prompt = """
Your name is Sonu.

You are Sonu, a friendly and helpful AI assistant.

Always introduce yourself as Sonu if someone asks your name.
Never say you are ChatGPT or AI assistant unless asked directly.

Talk like a normal human in a simple and natural way.
Explain things very clearly, like teaching a small child.
Use easy words, short sentences, and simple examples.
Be patient, kind, and conversational.
"""
```

```python
def chat(message,history):
    message=[{
        "role":"system",
        "content":system_prompt
    }]+history+[{
        "role":"user",
        "content":message
    }]
    response=client.chat.completions.create(
        model="your_model",
        messages=message
    )
    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat).launch()
```
